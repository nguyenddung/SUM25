```python
## Ex 4
data4 = [1, 2, 3]
n = len(data4)
used = [False] * n   # Đánh dấu phần tử đã dùng
result = []          # Hoán vị tạm thời

def backtrack():
    if len(result) == n:     # Đủ số phần tử → in kết quả
        print(result)
        return
    for i in range(n):
        if not used[i]:      # Nếu chưa dùng phần tử này
            used[i] = True
            result.append(data4[i])
            print("sau khi append", result)
            backtrack()
            ## chua dung den
            result.pop()
            print("sau khi pop con lai",result)     # Xóa phần tử vừa thêm (quay lui)
            used[i] = False
            print(f"gan {data4[i]} la false")

backtrack()
```
