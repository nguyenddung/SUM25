```python
import sqlite3
import pandas as pd

# Load CSV data
df = pd.read_csv('3. StudentsPerformance.csv')

# Connect to SQLite database (or create it)
conn = sqlite3.connect('mytable.db')
cursor = conn.cursor()

# Save DataFrame to SQL table
df.to_sql('students_performance', conn, if_exists='replace', index=False)

# Read the table back from the database
# tables = pd.read_sql("SELECT * FROM students_performance WHERE gender = 'female'", conn)
table2 = pd.read_sql("SELECT id,race,[parental level of education] FROM students_performance",conn)
# Display the DataFrame
print(table2)

# Close the connection
```
