#### hàm calcHist quan trọng để vẽ biểu đồ histgram
biểu đồ histogram là đếm số lượng kênh màu

```python
colors = ('b', 'g', 'r')
for i, color in enumerate(colors):
    hist = cv2.calcHist([img_rgb], [i], None, [256], [0,256])
    plt.plot(hist, color=color)
plt.title('Histogram các kênh màu')
plt.xlabel('Giá trị pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
count = np.zeros(256)
for i in range(img_rgb.shape[0]):
    for j in range(img_rgb.shape[1]):
        value = img_rgb[i][j]
        count[value[0]] += 1
```

```python
plt.plot(count)
plt.plot(count, color='r')
plt.show()
```

```python
count = np.zeros(256)
count1 = np.zeros(256)
count2 = np.zeros(256)
for i in range(img_rgb.shape[0]):
    for j in range(img_rgb.shape[1]):
        count[img_rgb[i][j][0]] += 1
        count1[img_rgb[i][j][1]] += 1
        count2[img_rgb[i][j][2]] += 1
plt.plot(count, color='b')
plt.plot(count1, color='g')
plt.plot(count2, color='r')
plt.show()
```

```python
hist = cv2.calcHist([img_rgb], [0], None, [256], [0, 256])
calmuter = np.zeros_like(hist)
for i in range(len(calmuter)):
    calmuter[i] = hist[:i].sum()
plt.plot(calmuter, color='b')
plt.show()
```

```python
img_eq = img_rgb.copy()
for i in range(img_eq.shaep[0]):
    for j in range(img_eq.shape[1]):
        img_eq[i][j][0] = calmuter[img_rgb[i][j][0]]
        img_eq[i][j][1] = calmuter[img_rgb[i][j][1]]
        img_eq[i][j][2] = calmuter[img_rgb[i][j][2]]
```
