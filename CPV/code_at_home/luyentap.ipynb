```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Tọa độ tam giác (3 đỉnh + quay về điểm đầu để vẽ kín)
triangle = np.array([
    [0, 1, 1, 0],   # x
    [0, 0, 1, 0],     # y
    [1, 1, 1, 1]      # homogeneous coordinate
])
```

```python
plt.grid(True)
plt.xlim(0,5)
plt.ylim(0,5)
plt.plot(triangle[0],triangle[1],marker = 'd', lw = 2)  # Vẽ tam giác
```



#### Sẽ nối A → B → C nhưng không nối lại C → A, nên hình sẽ mở.

✅ Vì thế, để vẽ tam giác khép kín (A → B → C → A), ta phải lặp lại đỉnh A ở cuối

triangle = np.array([

```python
# Các điểm A, B, C, rồi lặp lại A để vẽ khép kín tam giác
triangle2 = np.array([
    [0,   2,   0.5,  0],   # x
    [0, 0.5,   3,    0],   # y
    [1,   1,   1,    1]    # tọa độ đồng nhất (homogeneous)
])

plt.plot(triangle2[0],triangle2[1],marker = 'o', lw = 2)  # Vẽ tam giác
```

```python
hcn = np.array([
    [0, 2, 2, 0, 0],   #x
    [0,0,2,2,0],    #y
    [1,1,1,1,1]     # tọa độ đồng nhất (homogeneous)
])
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid(True)
plt.plot(hcn[0],hcn[1],marker = 'o', lw = 2)  # Vẽ hình chữ nhật
```

```python
# Tịnh tiến (2,1)
T = np.array([
    [1, 0,2],  # cộng 2 vào x
    [0, 1, 1],  # cộng 1 vào y
    [0, 0, 1]
])
```

```python
hcn_dich = T @ hcn
plt.xlim(0,10)
plt.ylim(0,10)
plt.grid(True)
plt.plot(hcn[0],hcn[1],marker = 'o', lw = 2)  # Vẽ hình chữ nhật
plt.plot(hcn_dich[0],hcn_dich[1],marker = 'o', lw = 2)  # Vẽ hình chữ nhật
```
