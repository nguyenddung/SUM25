```python
import numpy as np
tx = 2
ty = 3
T = np.array([[1, 0, tx],
              [0, 1, ty],
              [0, 0, 1]])
sx = 2
sy = 1
S = np.array([[sx, 0, 0],
              [0, sy, 0],
              [0, 0, 1]])


A = np.array([1, 1])
B = np.array([1, 2])
C = np.array([0, 1])
G = np.array([A, B, C, A])
# print(G)
one = np.ones((G.shape[0], 1))  
Gh = np.concatenate([G, one], axis=1)
# print(Gh)
```

```python
Gt = np.zeros_like(Gh)
Gs = Gt.copy()
for i in range(Gh.shape[0]):
    Gt[i] = T.dot(Gh[i])
    Gs[i] = S.dot(Gh[i])
```

```python
import matplotlib.pyplot as plt
plt.plot(G[:, 0], G[:, 1],marker ='d',lw = 2)
plt.plot(Gt[:, 0], Gt[:, 1],marker ='o',lw = 2)
plt.plot(Gs[:, 0], Gs[:, 1],marker ='x',lw = 2)
plt.xlim(0,4)
plt.ylim(0,8)
plt.grid(True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Bc 1: nh ngha cc nh ca tam gic theo ta  ng nht.
# Chn cc nh: A(0, 0), B(1, 0), C(0.5, 1).  v ng bao khp kn, ta lp li nh A  cui.
triangle = np.array([
    [0,   1, 0.5, 0],   # Ta  x
    [0,   0,   1, 0],   # Ta  y
    [1,   1,   1, 1]    # Phn t ng nht (w lun = 1)
])

# Bc 2: nh ngha cc ma trn bin i 3x3.
# 2.1: Ma trn Co gin (Scaling) vi h s Sx=1.2 v Sy=0.8.
S = np.array([
    [1.2, 0,   0],
    [0,   0.8, 0],
    [0,   0,   1]
])

# 2.2: Ma trn Xoay (Rotation) vi gc 30 (chuyn v radian).
theta = np.deg2rad(30)
R = np.array([
    [np.cos(theta), -np.sin(theta), 0],
    [np.sin(theta),  np.cos(theta), 0],
    [0,              0,             1]
])

# 2.3: Ma trn Tnh tin (Translation) dch chuyn theo (tx=2, ty=1.5).
T = np.array([
    [1, 0, 2],
    [0, 1, 1.5],
    [0, 0, 1]
])

# Bc 3: Kt hp cc php bin i.
# Th t thc hin: Co gin  Xoay  Tnh tin.
Transform = T @ R @ S  # T * (R * S)

# Bc 4: p dng ma trn bin i cho tam gic.
transformed_triangle = Transform @ triangle

# Chuyn i t ta  ng nht v ta  thng bng cch chia cho thnh phn th 3.
transformed_triangle = transformed_triangle / transformed_triangle[2, :]

# Bc 5: V hnh tam gic gc v tam gic sau bin i.
plt.figure(figsize=(8, 8))
plt.plot(triangle[0, :], triangle[1, :], 'bo-', label='Tam gic gc')
plt.plot(transformed_triangle[0, :], transformed_triangle[1, :], 'ro-', label='Tam gic bin i')
plt.title("Bin i Affine trn Hnh Tam Gic")
plt.xlabel("Trc X")
plt.ylabel("Trc Y")
plt.legend()
plt.axis('equal')
plt.grid(True)
plt.show()
```
